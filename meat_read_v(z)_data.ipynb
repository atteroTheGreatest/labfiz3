#Dane z charakterystykami V(z) dla mięsa

```python
import pandas as pd
```

```python
def read_data(filename):
    """Read integer data from file.
    
    Parameters:
    :filename - name of file to read data from
    
    It will ommit first 3 lines of input,
    because in our case, those lines doesn't
    represent our data.
    
    Return data in pandas Series."""
    readings = []
    with open(filename) as f:
        for line in f.readlines()[3:]:
            readings.append(int(line.strip()))
    return pd.Series(readings)
```

##Example plot of data

```python
m = read_data('data/3MIES_V1')
m.plot()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(14, 8))
plt.plot(m.index, m, 'o-')
plt.xlim(340, 800)
```

```python
def plot_series_data(data, title):
    plt.figure(figsize=(14, 8))
    plt.plot(data.index, data, '-')
    xmin = m[m >= m.max()].index[len(m[m >= m.max]) / 5]
    xmax = m[m <= m.min()].index[-10]
    plt.xlim(xmin, xmax)
    plt.title(title)
```

```python
import os
```

```python
basedir = 'data'
datafilenames = [basedir + '/' + filename for filename in os.listdir(basedir) if 'V' in filename]
```

```python
for filename in datafilenames:
    data = read_data(filename)
    plot_series_data(data, filename)
```

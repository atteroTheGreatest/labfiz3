```python

def describe_contours(contours):
    contours_count = len(contours)
    print("Number of detected contours:", contours_count)
    lengths = np.array([contour_length(contour) for contour in contours])

    average_length = lengths.mean()
    
    # cut off outliers
    lengths_cut = lengths[lengths < average_length + 20]
    variance = lengths_cut.var()
    print("Mean:", lengths_cut.mean())
    print("Variance", variance)
    print("Median", np.median(lengths))
    plt.figure(figsize=(12, 6))
    plt.hist(lengths_cut, bins=20)
    plt.title(u'$Długości\  konturów$', fontsize=20)
```

```python
def contour_length(contour):
    length = 0
    for x1, x2 in zip(contour[:], contour[1:]):
        length += np.sqrt((x1[0] - x2[0])**2 + (x1[1] - x2[1])**2)
    return length
```
